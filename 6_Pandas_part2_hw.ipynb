## Задание 3.4

Определите, в какие дни недели лучше покупать евро, а в какие – продавать. Для этого рассчитайте среднюю доходность евро в разные дни недели (т.е. нужно сгруппировать таблицу по дням недели и рассчитать среднее процентное изменение курса евро в каждой группе).

```python
days = {
    6:{'Weekday':u'воскресенье'},
    0:{'Weekday':u'понедельник'},
    1:{'Weekday':u'вторник'},
    2:{'Weekday':u'среда'},
    3:{'Weekday':u'четверг'},
    4:{'Weekday':u'пятница'},
    5:{'Weekday':u'суббота'}
}
dfdays = pd.DataFrame.from_dict(days,orient='index')

dfnew = pd.concat([df,(df.EUR-df.shift().EUR).rename("EUR_DIFF")],axis=1)

dfnew['weekdaynum'] = dfnew.Date.dt.weekday 
dfnew = pd.merge(dfnew, dfdays, how='left', left_on='weekdaynum', right_index=True).drop('weekdaynum',axis=1)


dfmean = dfnew[['Date','EUR','EUR_DIFF','Weekday']].groupby('Weekday').mean()

dfmean.EUR_DIFF/dfmean.EUR*100
```
